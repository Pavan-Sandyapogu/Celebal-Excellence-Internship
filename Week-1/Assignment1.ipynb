```python
# Step3: Handle missing values (identify, fill/drop)
```

```python
df.isnull().sum()
```

```python
 # we will fill columns like discount , seller_name, seller_information where as we will drop columns like 
 # what_customers_said, videos, variations(since data available is less than 50 %)
```

```python
df['discount']=df['discount'].fillna(df['discount'].mean())
```

```python
df['seller_name']=df['seller_name'].fillna('Unknown Seller')
```

```python
df['seller_information']=df['seller_information'].fillna('Un Available')
```

```python
df=df.drop(columns=['what_customers_said','videos','variations'])
```

```python
# After dropping and filling Missing values in the DataFrame
```

```python
print("No.of Missing Values in the DataFrame are : \n",df.isnull().sum())
```

```python
print("The DataFrame Available is :")
df
```

```python
# Step 4: Perform basic operations (filter rows, select columns)
```

```python
#Filtering Rows based on Seller_name
df[df['seller_name']!='Unknown Seller']
```

```python
#Filtering Rows based on Discount(i.e discount>=75)
df[df['discount']>=75]
```

```python
# Filtering columns(i.e Selecting Only Required Columns)
filtered_df=df[['product_id','title','category','initial_price','discount','final_price']]
filtered_df
```

```python
# Now We try to filter the same columns but only the rows which has discount>=75
filtered_df2=filtered_df[filtered_df['discount']>=75]
filtered_df2
```

```python
# Step 5:Remove duplicates.
```

```python
print("No.of Duplicate Rows available in the DataFrame are : ")
df.duplicated().sum()
```

```python
# Step 6: Create a derived column
```

```python
# Creating/deriving  a column money_saved, it will indicate amount of money saved by ordering this product
df['money_saved']=df['initial_price']*(df['discount']/100)
```
